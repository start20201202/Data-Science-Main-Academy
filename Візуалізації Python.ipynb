```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
m1 = 0.5
sigma1 = 0.3

m2 = 1.5
sigma2 = 0.3

samples1 = np.random.normal(m1, sigma1, 30000)
samples2 = np.random.normal(m2, sigma2, 30000)
```

```python
samples = np.concatenate([samples1, samples2])
```

```python
plt.hist(samples, bins=30)
plt.show()
```

```python
df = pd.DataFrame({
    'date': pd.to_datetime(['2023-01-01', '2023-01-02', '2023-01-03', '2023-01-04', '2023-01-05', '2023-01-01']),
    'value': [10, 20, 30, 10, 50, 5]
})
```

```python
df = pd.DataFrame({
    'height': np.random.randint(150, 200, 100),
    'weight': np.random.randint(40, 120, 100),
    'gender': np.random.choice(['male', 'female'], 100)
})
```

```python
df
```

```python
fig, ax = plt.subplots(1, 2, figsize=(20,5))
sns.boxplot(data=df, x='height', ax=ax[0])
sns.boxplot(data=df, x='weight', ax=ax[1])
# sns.boxplot(data=df, x='gender', ax=ax[2])
ax[0].set_title('Height')
ax[1].set_title('Weight')
plt.show()
```

```python
df = pd.DataFrame({
    'country': ['USA', 'UA', 'UK', 'FR', 'GR'],
    'population': [100, 70, 60, 65, 59]
})
```

```python
df
```

```python
plt.figure(figsize=(5, 5))
plt.pie(df.population, labels=df.country, autopct='%1.1f%%')
plt.show()
```
